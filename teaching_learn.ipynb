```python
TREE_STYLE = """
<style>
.tree-chapter {
    margin: 20px 0 8px 0;
    padding: 10px 20px;
    background: linear-gradient(90deg, #4a5568 0%, #718096 100%);
    color: white;
    border-radius: 8px;
    font-size: 17px;
    font-weight: bold;
}
.tree-table {
    width: 100%;
    border-collapse: separate;
    border-spacing: 0;
    margin: 0 0 10px 0;
    border-radius: 8px;
    overflow: hidden;
    box-shadow: 0 1px 4px rgba(0,0,0,0.08);
}
.tree-table th {
    background: #edf2f7;
    color: #4a5568;
    padding: 8px 14px;
    font-size: 13px;
    text-align: center;
    border-bottom: 2px solid #cbd5e0;
}
.tree-table td {
    padding: 7px 14px;
    font-size: 14px;
    border-bottom: 1px solid #edf2f7;
    vertical-align: middle;
}
.tree-section {
    background: #f7fafc;
    font-weight: bold;
    color: #2d3748;
    font-size: 14px;
}
.tree-num {
    text-align: center;
    font-weight: bold;
    color: #4a90d9;
    font-size: 15px;
}
.tree-concept { text-align: center; }
.tree-exercise { text-align: center; font-size: 13px; }
.tree-has { color: #38a169; }
.tree-none { color: #cbd5e0; }
</style>
"""

def show_knowledge_tree():
    """展示完整知识树，带编号方便选择"""
    topic_index = {}
    idx = 1

    display(HTML(TREE_STYLE))

    for ch_id, ch in KNOWLEDGE_TREE.items():
        display(HTML(f'<div class="tree-chapter">第{ch_id}章 &nbsp;{ch["title"]}</div>'))

        rows = ''
        for sec_id, sec in ch['sections'].items():
            rows += f'<tr><td class="tree-section" colspan="4">{sec_id} {sec["title"]}</td></tr>\n'
            for topic in sec['topics']:
                has_c = '<span class="tree-has">📖 有</span>' if topic in CONCEPTS else '<span class="tree-none">—</span>'
                ex_count = len(EXERCISES.get(topic, []))
                has_e = f'<span class="tree-has">{ex_count} 题</span>' if ex_count > 0 else '<span class="tree-none">—</span>'
                rows += f"""<tr>
                  <td class="tree-num">{idx}</td>
                  <td>{topic}</td>
                  <td class="tree-concept">{has_c}</td>
                  <td class="tree-exercise">{has_e}</td>
                </tr>\n"""
                topic_index[idx] = topic
                idx += 1

        display(HTML(f"""
        <table class="tree-table">
          <thead><tr><th style="width:60px">编号</th><th>知识点</th><th style="width:70px">概念</th><th style="width:70px">练习</th></tr></thead>
          <tbody>{rows}</tbody>
        </table>"""))

    return topic_index

topic_index = show_knowledge_tree()
print(f'\n共 {len(topic_index)} 个知识点。请在下方输入编号学习对应知识点。')
```
